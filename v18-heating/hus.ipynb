# Uppvärmning av hus

Den här uppgiften går ut på att skapa en beräkningsmodell i Jupyter för att simulera uppvärmningsbehovet för ett (kraftigt förenklat) hus. 

*Uppgiften går att lösa på olika nivå / svårighetsgrad.* Exempelvis kan utomhustemperaturen beskrivas på något av följande sätt:
- 1 årsmedeltemperatur (kraftig förenkling)
- 2-4 säsongstemperaturer (sommar, vinter)
- 12 månadsmedeltemperaturer i en lista
- tim / dygnsdata för ett år (t.ex. 2019) från en riktig datakälla (*Psss... SMHI!*)

Uppgiften löses med fördel **iterativt**, dvs <mark>börja med att göra en mycket enkel modell</mark>. Öka sedan komplexiteten stegvis. Försök inte göra allt från scratch. Ta inte vatten över huvudet.

## Instruktioner
Du har 2 lektionstillfällen att arbeta med modellen. Du får naturligtvis även jobba med modellen hemma om du vill, men det är inget som förväntas om du utnyttjar lektionstiden väl.

### Huset
- Huset är ett rätblock med en viss bredd, längd och höjd
- Huset saknar vind
- Huset saknar fönster och dörr
- Huset har perfekt temperaturreglering, dvs. **den har en värmekälla som i varje stund tillgodoser uppvärmningsbehovet** vilket leder till att inomhustemperaturen alltid är samma
  - *Grundfrågan i hela uppgiften är: Vilken effekt måste värmekällan ha för att kunna åstadkomma det här? Hur mycket energi blir det på ett år?*

# Bedömning
Krav för godkänd:
- Modellen klarar av att beräkna en årlig energimängd (**svara i MWh**) för uppvärmning utifrån
